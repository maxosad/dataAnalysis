```python
import pandas as pd
import numpy as np


df = pd.read_csv('myNba2k20-full.csv')
```

```python
def eraseDollar(value: object ) -> int:
    return int(value.replace("$", ""))
```

```python
k76 = 0
k86 = 0 
k97 = 0
def grp(x):
    global k76
    global k86
    global k97
    
    if (x[2] <= 76): 
        k76 = k76 + 1
        return 76
    elif (x [2] <= 86):
        k86 = k86 + 1
        return 86
    else: 
        k97 = k97 + 1
        return 97
```

```python
df.rating = df.apply(grp, axis = 1)
```

```python
print(k76)
print(k86)
print(k97)
```

```python
def h(x) -> float:
    #a = float(x[6].split('/ ')[1])
    a = x[6].split('/ ')[1]
    return float(a)
```

```python
df.height = df.apply(h, axis = 1)
```

```python
def w(x: object) -> float:
    a = x[7].split('/ ')[1]
    b = a.split()[0]
    return float(b)
```

```python
df.weight = df.apply(w, axis = 1)
```

```python
df['weightLog'] = np.log(df.weight)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
knn_clf = KNeighborsClassifier()
svm_clf = SVC()
forest_clf = RandomForestClassifier()
```

```python
knn_acc_score = cross_val_score(knn_clf, df[['height', 'weightLog']], df.rating, scoring='balanced_accuracy').mean()
svm_acc_score = cross_val_score(svm_clf, df[['height', 'weightLog']], df.rating, scoring='balanced_accuracy').mean()
forest_acc_score = cross_val_score(forest_clf, df[['height', 'weightLog']], df.rating, scoring='balanced_accuracy').mean()
```

```python
leaderboard = sorted([
    ('KNN', knn_acc_score),
    ('SVC', svm_acc_score),
    ('Forest', forest_acc_score)
], key=lambda pair: pair[1], reverse=True)

print('Leaderboard:')
for idx, (leader_name, leader_score) in enumerate(leaderboard):
    print(f'{idx + 1}. {leader_name}: {leader_score}')
```
